# The graph

What’s important here is that the graph must be created outside of the layout; our layout only calls for the finished visualization. Therefore, we place the chart code before the layout code

```python
import plotly.graph_objects as go
import numpy as np
import random
```

```python
fig = go.Figure(
    go.Scattergl(
        x=np.random.randn(1000),
        y=np.random.randn(1000),
        mode='markers',
        marker=dict(
            color=random.sample(
                ['#ecf0c1'] * 500 + ["#9400aa"] * 500, 1000),
            line_width=1
        )
    )
)

fig.update_layout(
    plot_bgcolor='#010103',
    width=790,
    height=730,
    xaxis_visible=False,
    yaxis_visible=False,
    showlegend=False,
    margin=dict(l=0, r=0, t=0, b=0)
)
```
